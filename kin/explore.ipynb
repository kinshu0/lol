```python
platforms = ['facebook', 'whatsapp', 'messages', 'instagram', 'slack', 'twitter']
platforms_split = [20, 100, 40, 30, 80, 15]

conversation_ids = conversations['conversation_id'].unique()

platforms_df = pd.DataFrame({'conversation_id': conversation_ids, 'platform': np.random.choice(platforms, len(conversation_ids), 
p=platforms_split/np.sum(platforms_split))})
# join platforms to conversations
# conversations = conversations.merge(platforms_df, on='conversation_id')
```

```python
people = pd.read_csv('us-500.csv')
```

```python
# pick 50 random people
people = people.sample(150)
```

```python
people['id'] = people.index
```

```python
people
```

```python
# assign conversation ids to person id based on benford's law
platforms_df['person_id'] = np.random.choice(people['id'], len(conversation_ids), p=people['id']/np.sum(people['id']))
```

```python
platforms_df.head()
```

```python
# join platforms to conversations
conversations = conversations.merge(platforms_df, on='conversation_id')
```

```python
platforms_df
```

```python
# bar chart breakdown of platforms
sns.countplot(x='platform', data=conversations)
```

```python
# plot frequency of person id
sns.countplot(x='person_id', data=platforms_df)
```

```python
# join people to conversations
conversations = conversations.merge(people[['first_name', 'last_name', 'phone1', 'id']], left_on='person_id', right_on='id')
```

```python
conversations.to_csv('conversations.csv', index=False)
```
